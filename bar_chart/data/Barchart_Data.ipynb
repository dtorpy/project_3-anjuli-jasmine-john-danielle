```python
import pandas as pd
```

```python
# Create a reference the CSV file desired
csv_path = "Data\Texas_2012.csv"

# Read the CSV into a Pandas DataFrame
wells_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
wells_df.head()
```

```python
well_count = wells_df['OperatorName'].value_counts()
df=pd.DataFrame(well_count)
well_count=df.reset_index()
well_count.head()
```

```python
well_count.to_csv('Data/well_count_BarChart.csv',index = False, header = False)
```
