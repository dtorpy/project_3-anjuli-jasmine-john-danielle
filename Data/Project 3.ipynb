```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
import warnings
warnings.filterwarnings('ignore')
import datetime as dt
from datetime import timedelta
from datetime import datetime
```

```python
#Pull CSV Files
registry1 = "registryupload_1.csv"
registry2 = "registryupload_2.csv"
registry3 = "registryupload_3.csv"
```

```python
#Read CSV Files
registry1_df = pd.read_csv(registry1)
registry2_df = pd.read_csv(registry2)
registry3_df = pd.read_csv(registry3)
```

```python
#Merge dataset 1 & 2
reg1_reg2 = registry1_df.merge(registry1_df, how='outer')
reg1_reg2
```

```python
#Merge 1/2 to dataset 3
total_reg = reg1_reg2.merge(registry3_df, how='outer')
total_reg                
```

```python
#Drop unecessary columns
total_reg.drop(columns =['pKey','JobStartDate','APINumber','StateNumber','CountyNumber','Projection', 'TVD', 'TotalBaseNonWaterVolume', 'FFVersion', 'FederalWell', 'IndianWell', 'Source', 'DTMOD',], inplace = True)
total_reg
```

```python
#Drop NA values
total_reg.dropna(subset = ['TotalBaseWaterVolume'], inplace = True)
total_reg
```

```python
#See type of each column
total_reg.dtypes
```

```python
#Change date column from 'object' to 'datetime'
total_reg['JobEndDate'] = pd.to_datetime(total_reg['JobEndDate'])
```

```python
total_reg.dtypes
```

```python
#Show only year of the date column
total_reg['JobEndDate'] = pd.to_datetime(total_reg['JobEndDate']).dt.year
total_reg
```

```python
texas_wells = total_reg.loc[total_reg["StateName"]== "Texas"]
texas_wells
```

```python
#Put new dataframe into a CSV file
texas_wells.to_csv('Texas_2012_2022.csv',index = False, header = True)
```
